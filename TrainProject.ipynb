# Apply returns some value after passing each row/column of a data frame with some function.
## The function can be both default or user-defined. For instance here it can be used to find the missing values in each row and column.
## axis=0 defines that function is to be applied on each column
## axis=1 defines that function is to be applied on each row

```python
#Create a new function:
def ValuesMissing(x):
  return sum(x.isnull())

#Applying per column:
print("Missing values per column:")
print(data.apply(ValuesMissing, axis=0)) 

#Applying per row:
print("Missing values per row:")
print(data.apply(ValuesMissing, axis=1).head()) 
```
